```python
shop['datetime']=pd.to_datetime(shop['datetime'])
shop['date']=pd.to_datetime(shop['date'])
```

```python
shop['year'] =shop['datetime'].dt.year
shop['month'] =shop['datetime'].dt.month_name()
shop['day'] =shop['datetime'].dt.day
shop['hour'] =shop['datetime'].dt.hour
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y='coffee_name',data=shop,order=shop['coffee_name'].value_counts().index)
plt.title("Most Popular Coffe Types")
plt.show
```

```python
plt.figure(figsize=(6,4))
sns.histplot(shop['money'])
plt.title("Distribution of Transaction Amounts")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='coffee_name',y='money',data=shop)
plt.xticks(rotation=45)
plt.title("Price Distribution by Coffee Type")
plt.show()
```

```python
daily_sales= shop.groupby('date')['money'].sum().reset_index()

plt.figure(figsize=(12,5))
plt.plot(daily_sales['date'],daily_sales['money'])
plt.title("Daily sales over time ")
plt.xlabel("date")
plt.ylabel("total money spent")
plt.show()
```

```python
("Unique card IDs:",shop['card'].nunique())

top_customers=shop['card'].value_counts().head(10)
(top_customers)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='coffee_name', y='money', data=shop, palette='coolwarm')
plt.xticks(rotation=45)
plt.title("Price Distribution by Coffee Type")
plt.xlabel("Coffee Type")
plt.ylabel("Money (₹)")
plt.show()
```

```python
shop['hour']=shop['datetime'].dt.hour

plt.figure(figsize=(10,5))
sns.countplot(x='hour',data=shop, palette='magma')
plt.title("number of sales by hour of the day")
plt.xlabel("hour of day (24-hour)")
plt.ylabel("number of transaction")
plt.show
```

```python
top_cards = shop['card'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_cards.values ,y=top_cards.index,palette='cubehelix')
plt.title("top 10 frequent customers (by card ID)")
plt.xlabel("number of purchase")
plt.ylabel("card id")
plt.show()
```

```python
coffee_counts = shop['coffee_name'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(coffee_counts,labels=coffee_counts.index,autopct='%1.1f%%',startangle=90)
plt.title("Coffe type sales share")
plt.show
```

```python
In this graph show that latte coffee is most sale in the shop
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(shop.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
peak_hour = (
    shop.groupby(['coffee_name', 'hour'])
      .size()                       # Count number of sales per hour
      .reset_index(name='sales_count')
)
```

```python
peak_time_per_coffee = peak_hour.loc[peak_hour.groupby('coffee_name')['sales_count'].idxmax()]
plt.figure(figsize=(10,6))
sns.barplot(x='coffee_name', y='hour', data=peak_time_per_coffee, palette='viridis')
plt.title("Peak Sales Hour for Each Coffee Type")
plt.xlabel("Coffee Type")
plt.ylabel("Peak Hour of Sale (24-hour format)")
plt.xticks(rotation=45)
plt.show()
```
